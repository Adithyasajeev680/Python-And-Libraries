### MULTILEVEL INHERITANCE

```python
class animal:
  def speak(self):
    print("animal speaking")

class dog(animal):  # Child class (inheritance)
  def bark(self):
    print("dog barking")

class dogchild(dog):  #multilevel inheritance
  def eat(self):
    print("eating bread...")

d=dog()
d.bark()
d.speak()
# d.eat()  AttributeError: 'Dog' object has no attribute 'eat'

print()

d1=dogchild()  #multilevel child have all the method
d1.bark()
d1.speak()
d1.eat()

```

```python
class grandfather:
  def house(self):
    print("gradfather buy a house")
  def land(self):
    print("grand father buy a land")

class father(grandfather):
  def car(self):
    print("fathe bought a car")

class son(father):
  def bike(self):
    print("son bought a bike")

a=son()  #son can access both father and grandfather
a.house()
a.land()
a.car()
a.bike()
print()

b=father() #father can acess grandfather only
b.house()
b.land()
b.car()
print()

c=grandfather() #can acess it self only
c.house()
c.land()
```

### MULTIPLE INHERITANCE

```python
class calculation1:  #parent class
  def summation(self,a,b):
    print(a+b)

class calculation2:  #another parent class
  def multiplication(self,a,b):
    print(a*b)

class derived(calculation1,calculation2):  #child from that both parent classes
  def divide(self,a,b):
    print(a/b)

a=derived()
a.summation(10,40)
a.multiplication(10,5)
a.divide(100,2)
```

* If we want to make object from abstract class we can't directly make object from the abstract class.

* So we need to make child class for abstract class object making

* And also need to make same method ( function ) on child class also


* All child classes must impement the abstract methos of the parent abstract class.
* Abstract class can have both abstract methods and concrete ( regular ) methods

```python
from abc import ABC,abstractmethod

class shape(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

class circle(shape):
  def __init__(self,radius):
    self.radius=radius

  def calculate_area(self):
    print(f"circle area is : {3.14 *(self.radius)**2 }")

class rectangle(shape):
  def __init__(self,width,height):
    self.width=width
    self.height=height

  def calculate_area(self):
    print(f"rectangle area is : {self.width * self.height}")

c=circle(10)
r=rectangle(5,10)

c.calculate_area()
print()
r.calculate_area()
```

```python
import threading
import time

def download(file):
  print(f"downloading{file}")
  time.sleep(2)
  print(f"{file} downloaded")

files=["file1.txt","file2.txt","file3.txt"]
threads=[threading.Thread(target=download,args=(f,))for f in files]

for t in threads:
  t.start()

for t in threads:
  t.join()
```

```python
import threading
import time

# Function for the chef's work (Thread 1)
def chef_work():
  print("kitchen : starting work.")
  for i in range(5):
    print(f"chef: preparing dish {i+1}..")
    time.sleep(2)
    if i == 2:
      print("chef :calling assistant to wash utensils.")
      time.sleep(1)
      assistant_thread=threading.Thread(target=assistant_work)
      assistant_thread.start()
      assistant_thread.join()
      print("chef: assistant done .resuming dish preparation.")
  print("chef: all dishes are prepared.")

# Function for the assistant's work (Thread 2)
def assistant_work():
  for j in range(3):
    print(f" assistant : washing utensil {j+1}..")
    time.sleep(2)
  print(" assistant :all utensils are clean")

chef_thread=threading.Thread(target=chef_work)
chef_thread.start()
chef_thread.join()
print("kitchen : all work completed.")

```

# EXCEPTION HANDLING

## try-except block

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Please enter a valid number!")
else:
    print("Division successful! Result:", result)
finally:
    print("This block always runs.")
```

## else

```python
try:
  result =10/2
except ZeroDivisionError:
  print("Error")
else:
  print("Result:", result)
```

## finally

```python
try:
  file_= open("data.txt","r")

except FileNotFoundError:
  print("File not found!")

finally:
  print("End of program")
```

## raise exception

```python
try:
  age=int(input("enter yoor age :"))
  if age<0:
    raise ValueError("age cannot be negative")
  else:
    print(f"your age is {age}")
except ValueError as e:
  print(e)
print("try except finished")
```

## custom exception

```python
class InvalidAgeException(Exception):
  "raised when the input value is less than 18"
  pass
number=18
try:
  input_num=int(input("enter a number :"))
  if input_num< number:
    raise InvalidAgeException
  else:
    print("eligible to vote")

except InvalidAgeException:
  print("exception occured :invalid age")
```

# GENERATOR

GENERATOR-at least one yield (yield is to use pause)

```python
def greet():
  print("hi")
  yield 1
  print("how are you")
  yield 2
  print("are you there")
  yield 3
g=greet()
result=next(g)
print(result)
result=next(g)
print(result)
result=next(g)
print(result)
```
